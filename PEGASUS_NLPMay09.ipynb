###Extract reviews from google play store and convert into dataframe


```python
from google_play_scraper import app
import pandas as pd
import numpy as np
# US Market

from google_play_scraper import Sort, reviews_all


us_reviews = reviews_all(
    'com.omega.taxi',
    sleep_milliseconds=0,  # defaults to 0
    lang='en',  # defaults to 'en'
    country='kw',  # defaults to 'us'
    sort=Sort.NEWEST,  # defaults to Sort.MOST_RELEVANT
)

df_autoveh = pd.DataFrame(np.array(us_reviews), columns=['review'])  # Concatenated list (list of list)
df_autoveh = df_autoveh.join(pd.DataFrame(df_autoveh.pop('review').tolist()))  # to dataframe
print(type(us_reviews))
df_autoveh.head()
```

##import libraries

##Select the model

```python
model_name="google/pegasus-xsum"
```

##Initialize pretrained tokenizer

```python
pegasus_tokenizer = PegasusTokenizer.from_pretrained(model_name)
```

##Define PEGASUS Model

```python
pegasus_model=PegasusForConditionalGeneration.from_pretrained(model_name)
```

```python
text_in="The scale and sophistication of cyber threats continue to increase. Australians are increasingly reliant on the internet and the internet-connected devices we use daily. The digital economy is the future of Australia’s economy. Our experience with the COVID-19 pandemic demonstrated this by highlighting how much Australians interact and work online, trusting the internet for healthcare, working from home, education, entertainment and online shopping. It is more important than ever to protect Australians online from those who would do us harm."
```

#Create Tokens

```python
tokens=pegasus_tokenizer(text_in, truncation=True, padding="longest", return_tensors="pt")
#tokens=pegasus_tokenizer(ex_text, truncation=True, padding="longest", return_tensors="pt")
#tokens=pegasus_tokenizer(rev_list[0:24], truncation=True, padding="longest", return_tensors="pt")
tokens
```

```python
##Summarize text
encoded_summary = pegasus_model.generate(**tokens)
```

```python
##Decode summarized text
decode_summary = pegasus_tokenizer.decode(encoded_summary[0], skip_special_tokens=True)
```
